```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt

from utils.displayData import displayData
from utils.predict import predict
```

```python
print("Loading and Visualizing Data ...")

data = sio.loadmat("ex3data1.mat")
X = data["X"]
y = data["y"].reshape(-1)
y[y==10]=0

row,col = X.shape

rand_indices = np.random.permutation(row)
sel = X[rand_indices[:100]]

displayData(sel)
```

```python
print("Loading Saved Neural Network Parameters ...")
weights = sio.loadmat('ex3weights.mat')
print(weights.keys())

t1= weights["Theta1"]
t2= weights["Theta2"]
t2 = np.roll(t2,1, axis=0)
```

```python
p = predict(t1,t2,X)  
print("Accuracy: {:.3f}".format(np.mean(p==y)))
```

```python
perm = np.random.permutation(X.shape[0])

for i in range(len(perm)):
    print("Displaying Example Image")
    num = X[perm[i],:].reshape(1,400)
    
    displayData(num)
    pred = predict(t1, t2, num)
    print("Neural Network Prediction:{}".format(*pred))
    plt.show()
    
    choice = input("Paused - press enter to continue, q to exit:")
    if choice == "q":
        break
```
